* 사망자수 column을 제외하고 모두 없애세요 : dead

```python
dead = complete_lifetable['사망자수']

# print head of dead.
print(dead.head())
```

#### **설명**
이 설명들은 허명회 교수님의 '법과 통계학' 교재의 95페이지, 96페이지의 내용입니다.

- 연령 x ($\le 100$)
- 사망률 $q_x$ : 연령 x의 사람이 연령 x+1에 이르기 이전에 사망할 확률
- 생존자 수 $l_x$ : 출생인구 100,000명 중 연령 x에 이른 사람 수
- 사망자 수 $d_x$ : 연령 x에 생존, 연령 x+1에 이르지 못한 사람 수
- 기대여명 $e_x$ : 연령 x 이후 평균 생존시간

더 자세한 설명은 통계청의 자료를 참조하세요.

* 생존자 수 series를 만드세요. : live
    - _HINT : numpy에는 cumulative sum을 할 수 있는 method가 있습니다._

```python
# make live list.
live = np.ones(101)*100000
temp = list(dead.cumsum())[:-1]
temp.insert(0, 0)
live -= temp
print(live)
```

```python
# convert live list to series.
live = pd.Series(live, dtype = int)
live.index = dead.index

# print head of live.
print(live.head())
```

* 사망확률 series를 만드세요. : death_prob
    - 소숫점 밑 6번째 자리에서 반올림 하세요.

```python
death_prob = np.round(dead/live, decimals = 5)

# print head of death_prob.
print(death_prob.head())
```

* 정지인구 Series를 만드시오 : stop
    - x세 정지인구=[x세 생존자수+(x+1세 생존자수)]/2
    - 100+세의 정지인구는 5240입니다.
    - 반올림 하세요.
    - dtype : int

```python
stop = np.append(np.round((live.values[:100]+live.values[1:])/2), 5240)
stop = pd.Series(stop, dtype = int)
stop.index = dead.index

# print tail of stop.
print(stop.tail())
```

* 총생존년수 Series를 만드시오 : survived_year  
각 연령별 정지인구의 합 : $\sum_{x{\ge}i} 정지인구_{i}$

```python
survived_year = stop[::-1].cumsum()[::-1]

# print head of survived_year.
print(survived_year.head())
```

* 기대여명 Series를 만드시오. : expectancy  
총생존년수 / 생존자수
- 소숫점 아래 셋째 자리의 수에서 반올림하시오.

```python
expectancy = np.round(survived_year/live, decimals = 2)

# print tail of expectancy.
print(expectancy.tail())
```

#### Merge Series
지금까지 구한 모든 Series들을 합하여 Dataframe을 만들어 봅시다!  
- 연령 | 사망률 생존자수 사망자수 정지인구 총생존년수 기대여명

```python
zip1 = list(zip(death_prob, live, dead, stop, survived_year, expectancy))
lifetable = pd.DataFrame(zip1)
lifetable.index = dead.index
lifetable.columns = ['사망률', '생존자수', '사망자수', '정지인구', '총생존년수', '기대여명']

# display head of lifetable.
display(lifetable.head())

# display tail of lifetable.
display(lifetable.tail())
```
