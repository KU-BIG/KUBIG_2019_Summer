```python
import numpy as np
```

* seed를 70으로 설정하시오.

```python
np.random.seed(70)
```

* 초기값이 1000인 random_walk라는 리스트를 생성하시오.

```python
random_walk = [1000]
```

* for문을 이용해 다음과 같은 상황을 100번 반복하시오.
    - 도박꾼은 1000원을 갖고 게임을 시작한다.
    - 1부터 100까지의 룰렛을 돌린다.
    - 룰렛의 숫자가 10이하면 룰렛의 숫자만큼 돈을 얻는다.
    - 룰렛의 숫자가 10보다 크고 70이하면 5만큼 돈을 잃는다.
    - 룰렛의 숫자가 70보다 크고 95이하면 2만큼 돈을 얻는다.
    - 룰렛의 숫자가 95보다 크고 100이하면 10만큼 돈을 잃는다.
    - 0.001의 확률로 도박꾼은 룰렛을 돌렸을 때, 모든 돈을 잃는다.
    - 도박꾼이 가지고 있는 돈은 음수가 될 수 없다.
    - 100번의 도박이 모두 시행되거나 도박꾼이 가지고 있는 돈이 0이하가 되는 순간 도박은 종료된다.

```python
for x in range(100) :
        reserve = random_walk[-1]
        roulette = np.random.randint(1,101)
        if  roulette <=10 :
            reserve = reserve+roulette
        
        elif roulette <= 70:
            reserve = max(0, reserve - 5)
        
        elif roulette <= 95:
             reserve = reserve + 2
                
        else:
             reserve = max(0, reserve -10)
        
        if np.random.rand() <= 0.001 :
            reserve = 0
            
        if reserve <= 0:
            reserve = 0
            random_walk.append(reserve)
            break
            
        random_walk.append(reserve)
```

* 도박꾼이 가지고 있는 돈의 현황을 나타내는 그래프를 matplotlib를 이용해서 그리시오.

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk
plt.plot(random_walk)

# Show the plot
plt.show()
```

* 위와 같은 도박을 해서 도박꾼이 마지막에 도달하는 돈의 현황을 500번 관측하자.
* 한번 시뮬레이션이 끝날 때마다 시뮬레이션의 결과를 all_walks라는 리스트에 저장하시오.

```python
# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [1000]
    for x in range(100) :
        reserve = random_walk[-1]
        roulette = np.random.randint(1,101)
        if roulette <=10:
            reserve = reserve+roulette
        
        elif roulette <= 70:
            reserve = max(0, reserve - 5)
        
        elif roulette <= 95:
             reserve = reserve + 2
                
        else:
             reserve = max(0, reserve -10)
        
        if np.random.rand() <= 0.001 :
            reserve = 0
            
        if reserve <= 0 :
            reserve = 0
            random_walk.append(reserve)
            all_walks.append(random_walk[-1])
            break
            
        random_walk.append(reserve)
    all_walks.append(random_walk[-1])
```

* all_walks라는 리스트를 numpy array로 변환하고 ends라는 변수에 저장하시오.

```python
ends = np.array(all_walks)

# display plot
plt.plot(ends)
plt.show()
```

```python
# ends의 histogram을 그리시오.
plt.clf()
plt.hist(ends)
plt.plot()
```

* dictionary인 id의 key와 value를 사용해 for loop을 작성하고
    - "x의 학번은 y입니다."를 출력하세요.

```python
id={'2011123':'영희','2011456':'민수','2011789':'민식','2011321':'영구'}

for key,value in id.items():
    print(value+"의 학번은"+" "+key+"입니다.")
```
