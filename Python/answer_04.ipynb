```python

##The investment opportunity set은 모든 가능한 risk-return 조합들로 이루어져 있는 집합이다.
##자산 A와 자산 B가 있다.
#자산 A는 expected return이 0.136이다. i.e. E[Ra]=0.136
#A의 return의 standard deviation은 0.154이다. i.e. Sigma_a=0.154
#자산 B는 expected return이 0.15이다. i.e. E[Rb]=0.15
#B의 return의 standard deviation은 0.23이다. i.e. Sigma_b=0.23
#w를 자산 A에 투자한 비율이라고 하자. W는 0과 1사이의 값으로 선택될 수 있다.



import numpy as np
import matplotlib.pyplot as plt


##list comprehension을 이용해 자산 A와 B로만 이루어진 포트폴리오의 
##기대수익률의 기대값을 성분으로 하는 리스트를 만들고 erp라는 변수에 저장하시오.
#이 때 w를 0부터 0.01씩 증가해 1에 이르는 sequence를 사용하시오.
#즉, iterator variable은 w입니다.

erp=[[w*0.136+(1-w)*0.150] for w in np.arange(0,1,0.01)]

##list comprehension을 이용해 자산 A와 B로만 이루어진 포트폴리오의
##기대수익률의 분산을 성분으로 하는 리스트를 만들고 vrp라는 변수에 저장하시오.
#이 때 w를 0부터 0.01씩 증가해 1에 이르는 sequence를 사용하시오.
#즉, iterator variable은 w입니다.

vrp=[[w**2*0.154**2+(1-w)**2*0.230**2+2*w*(1-w)*(0.154)*(0.230)*0.27] for w in np.arange(0,1,0.01)]



plt.plot(vrp,erp)
plt.title("investment opportunity set")
plt.xlabel("standad deviation")
plt.ylabel("expected return")
plt.show()

```
